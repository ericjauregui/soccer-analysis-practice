```python
pm["ob_touches"] = pm["throw"] + pm["corner"]

pm["player_desc"] = pm["name"] + " is the " + pm["team"] + " " + pm["pos"]

pm["at_least_one_throwin"] = pm["throw"] > 0

pm["len_last_name"] = pm["name"].apply(lambda x: len(x.split(" ")[-1]))

pm["match_id"] = pm["match_id"].astype(str)

pm["air_duel_won_pct"] = pm["air_duel_won"] / pm["air_duel"]

pm["air_duel_won_pct"] = pm["air_duel_won_pct"].fillna(-99)

pm.drop("air_duel_won_pct", axis=1, inplace=True)

pm.head()
```
